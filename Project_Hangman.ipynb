this is a simple python text game that stores a word and asks you to guess its letters. You win if all letters are guessed correctly. You lose if your correct guesses are not up to the number of letters the word is made with. Ohhh! remember, each time you guess wrongly, an extra feature of Hangman is printed.....know that you are on your way to hang the man...Anyways! Let's dive in.

```python

# Create Hangman characters in a list
Hangman = ['''
________
|      | 
|      o
|     \|/ 
|      |
|     / \\
_       ''',''''
________
|      | 
|      o
|     \|/  
|      |
|       
_       ''','''
________
|      | 
|      o
|     \|/ 
|
|      
_       ''', '''
________
|      | 
|      o
|      |/
|
|      
_       ''','''
________
|      | 
|      o
|      |
|      
|      
_       ''','''
________
|      | 
|      o
|      
|      
|     
_    

 ''','''
________
|      | 
|      
|      
|      
|     
_       ''']
# Welcome screen
print("Let's play Hangman!")
print('''
________
|      | 
|      
|      
|      
|     
_ ''')
print("I'm thinking of a word. Can you guess the letters that form it?")
print()
end_of_game = False
word = 'SENIOR'
word_length = len(word)
count = 6

# Create a variable called 'count' to keep track of the number of count left. 
# Set 'count' to equal 6.
# Create an empty list to store the correctly guessed letters
display_list = []
for _ in range(word_length):
    display_list += "_"

while not end_of_game:
    guess = input("Guess a letter: ").upper()
    print()
    #Check guessed letter
    for position in range(word_length):
        letter = word[position]
        if letter == guess:
          print('Good job, {} is in the word.'.format(letter))
          display_list[position] = letter

    # If guess is not a letter in the word, we 'count' by 1. 
    #If count goes down to 0 then the game ends and prints "You lose."
    print()
    if guess not in word :
      count -= 1
      print('{} is not in the word.'.format(guess))

      if count == 0 :
        end_of_game = True
        print()
        print("You lose.")
        
    # Join all the elements in the list and turn it into a String.
    print()
    print(' '.join(display_list))

    # Make sure user guessed all letters correctly.
    print()
    if "_" not in display_list :
      end_of_game = True
      print("You win.")

    # print the 'Hangman' that corresponds to the current number of 'counts' the user has remaining.
    
    print(Hangman[count]) 
print('..... Kindly brought to you by Aaron!')
```
